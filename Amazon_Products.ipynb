```python
#Import required libraries
import pandas as pd
```

```python
#Create a data frame in Pandas to manipiulate data
df=pd.read_csv("/content/amazon_product.csv")

#retrieve the first 5 rows of data
df.head()
```

```python
#Since Pandas automatically creates an index for every DataFrame, we deciced to delete the first column of the original data frame.
#remove the first column from data frame
df=df.drop("Unnamed: 0",axis=1)
df.head()
```

```python
#delete the dollar sign before the prices. Replace method replaces the values with new values.
#To change a charechter inside a value, we use ".str"
df['product_price']=df['product_price'].str.replace("$","")
df['product_original_price']=df['product_original_price'].str.replace("$","")
df["product_minimum_offer_price"]=df["product_minimum_offer_price"].str.replace("$","")
df[["product_price","product_original_price","product_minimum_offer_price"]]
```

```python
#change the data type of original price and price
df["product_original_price"]=df["product_original_price"].astype("float")
df["product_price"]=df["product_price"].astype("float")
df["product_minimum_offer_price"]=df["product_minimum_offer_price"].astype("float")
```

```python
#Delete the last 3 columns because of the number of Null values
df=df.drop(["product_availability","unit_price","unit_count"],axis=1)
df.head()

#check the numb of columns
df.shape
```

```python
#Determinig products which are best seller to do inventory management and highlt them in marketing campaings to attract customers
df_best_seller=df[df["is_best_seller"]==True]
```

```python
#create a new data frame with no null values in "product_original_price" column
df_cleaned_price=df.dropna(subset=["product_original_price"])
df_cleaned_price.shape
```

```python
#compare "product_original_price" and "product_price"
df_cleaned_price.plot(kind="bar",x="product_title",y=["product_price","product_original_price"])
```

```python
#Examine the distribution of product ratings
df_cleaned_star=df.dropna(subset="product_star_rating")
df_cleaned_star.shape
bin_edge=[0,1,2,3,4,5]
df_cleaned_star.plot(kind="hist",y="product_star_rating",bins=bin_edge,edgecolor="black")
#y determines the column that we want to plot its histogram
```

```python
#calculate the average of number of ratings for different star ratings provided by customers
#highly-rated and popular products generally receive more reviews. Approximately 60,000 customers gave near 5-star ratings to Amazon products.
g=df_cleaned_star[["product_star_rating","product_num_ratings"]].groupby("product_star_rating").mean()
g=g.reset_index()
g.dtypes
ax=g.plot(kind="scatter",x="product_star_rating",y="product_num_ratings")
ax.set_ylabel("Average number of product ratings")
ax.set_xlabel("star rating")
```
